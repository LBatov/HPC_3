```python
#читаю эксельники, собира полный датафрейм

from os import listdir
from os.path import isfile, join
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize
import numpy as np

mypath = "D:\\top500"
xlss = []
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
for f in onlyfiles:
    print(f)
    df = pd.read_excel(f"D:\\top500\\{f}", header=1)
    if "RPeak" not in df.columns:
        df = pd.read_excel(f"D:\\top500\\{f}", header=0)
    print(df.columns)
    for col in df.columns:
        if "TFlop" in col:
            df[col] = df[col] * 1000
    df.rename(
        columns={x: x.split(" ")[0].strip() for x in df.columns if "[" in x},
        inplace=True,
    )
    df.rename(columns={x: x.lower() for x in df.columns}, inplace=True)
    df["yyyymm"] = f.split("_")[1].split(".")[0]
    df["yyyymm_dt"] = datetime.strptime(f.split("_")[1].split(".")[0], '%Y%m')
    cc = sorted(df.columns, reverse=True)
    df = df[cc]
    print(df.columns)
    xlss.append(df)
    
    
full_df = pd.concat(xlss, axis=0)
full_df.columns
```

```python
# отбираю первые ранги за каждый временной отсчёт

top_perf_df = full_df[full_df['rank'] == 1][['yyyymm_dt','rmax']]
top_perf_df.head()
```

```python
# логарифмирую и подбираю коэффициенты прямой (пробовал экспонентой без логарифмирования - получается хуже)

x =  (top_perf_df['yyyymm_dt'].astype(np.int64).to_numpy())   
y =  np.log(top_perf_df['rmax'].to_numpy())

p=optimize.curve_fit(lambda t,a,b: a*t+b,  x,  y,  p0=(1000, 0.00001)) 
 

def lin(p,t): return p[0]*t+p[1]

plt.scatter(x,y)

check_date = datetime.strptime('202506', '%Y%m')
 
x=np.append(x, check_date.timestamp() * 1000000000)
plt.scatter(x,[lin(p[0], x) for x in x]) 

print("производительность в 2025 году -", round(np.e ** lin(p[0],x[-1]),2))
```

```python
# Нашел данные по потреблению энергии за 2021 год на https://yearbook.enerdata.net/electricity/world-electricity-production-statistics.html
TOTAL_PRODUCTION_WORLD_TWh = 27544

power_df = full_df[full_df['yyyymm'] == '202111']
power_df = power_df[power_df['power (kw)'].notnull()]
x =  power_df['rank'].to_numpy() 
#print(x[:10])
y =  power_df['power (kw)'].to_numpy()
 
```

```python
# восстановил пропуски при помощи обратной экспоненты (чисто визуально похоже)


def rev_exp_c(p, t):
    return 1 / (p[0] * np.exp(p[1] * t) + p[3]) + p[2]


p = optimize.curve_fit(
    lambda t, a, b, c, d: 1 / (a * np.exp(b * t) + d) + c,
    x,
    y,
    p0=(0.0001, 0.00001, 0.1, 0.0001),
)


plt.scatter(x, y, color="blue")
x_all = [x + 1 for x in range(500)]
y_all = np.array([rev_exp_c(p[0], x + 1) for x in range(500)])
plt.scatter(x_all, y_all, color="orange")
print("total consumption (kw)", round(np.sum(y_all),2))
```

```python

TOTAL_PRODUCTION_WORLD_KWh = 27544 * 1000 ** 3 / 365 / 24
super_cpu_percent =  np.sum(y_all)/TOTAL_PRODUCTION_WORLD_KWh*100
print('Процент энергии, которую потребили суперкомпьютеры', round(super_cpu_percent,3), '%')
```
